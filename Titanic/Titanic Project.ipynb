```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train_1=train.drop(['Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
```

```python
test_1=test.drop(['Name','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
```

```python
train_1.head()
```

```python
test_1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
train_1[["Sex"]] = encode.fit_transform(train_1[["Sex"]])
test_1[["Sex"]] = encode.fit_transform(test_1[["Sex"]])

#Fill '0' for null values of age column
train_1['Age'] = train_1['Age'].fillna(0)
test_1['Age'] = test_1['Age'].fillna(0)
train_1.head()
```

```python
train_1_features=train_1.drop(['PassengerId','Survived'],axis=1)
train_1_target=train_1['Survived']

test_1_features=test_1.drop(['PassengerId'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

LR_model=LogisticRegression()
LR_model.fit(train_1_features,train_1_target)
lr_predictions=LR_model.predict(test_1_features)

submission = pd.DataFrame({'PassengerId':test_1['PassengerId'],'Survived':lr_predictions})

submission.head()
filename = 'Titanic Predictions - Logistic Regression-1.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
```

```python
train_1_features['SibSp']=train['SibSp']
train_1_features['Parch']=train['Parch']
train_1_features.head()
test_1_features['SibSp']=test['SibSp']
test_1_features['Parch']=test['Parch']
test_1_features.head()
```

```python
LR_model_2=LogisticRegression()
LR_model_2.fit(train_1_features,train_1_target)
lr_predictions_2=LR_model_2.predict(test_1_features)

submission = pd.DataFrame({'PassengerId':test_1['PassengerId'],'Survived':lr_predictions_2})

submission.head()
filename = 'Titanic Predictions - Logistic Regression-2.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
```

```python
#Building Decision Tree Model
train_2_features=train_1_features.drop(['SibSp','Parch'],axis=1)
train_2_features.head()
test_2_features=test_1_features.drop(['SibSp','Parch'],axis=1)
test_2_features.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

DT_Model_1=DecisionTreeClassifier()
#maintaining same target variable as there are no changes
DT_Model_1.fit(train_2_features,train_1_target)
DT_Model_pred_1=DT_Model_1.predict(test_2_features)


submission = pd.DataFrame({'PassengerId':test_1['PassengerId'],'Survived':DT_Model_pred_1})

filename = 'Titanic Predictions - Decision Tree-1.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
submission.head()
```
