```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
x=df.drop('ACTION',axis=1)
y=df['ACTION']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

## 1) Logistic Regression Model


```python
Log_model=LogisticRegression()
Log_model.fit(x_train,y_train)
log_y_pred=Log_model.predict(x_test)
print("Logistic Regression accuracy score",accuracy_score(y_test,log_y_pred))

print("Confusion Matrix")
print(confusion_matrix(y_test,log_y_pred))
print("From Confusion matrix Model looks to be biased for label 1, So need to check with other models")
```

## 2) Random Forest Classifier

```python
random_forest_classifier=RandomForestClassifier()
random_forest_classifier.fit(x_train,y_train)
rfc_y_pred=random_forest_classifier.predict(x_test)

print("Random Forest Classifier accuracy score",accuracy_score(y_test,rfc_y_pred))
print("Confusion Matrix")
print(confusion_matrix(y_test,rfc_y_pred))

print("Accuracy score looks good and confusion matrix also shows there are good predictions for label 0")
```

## 3) Ada Boost Classifier

```python
ada_boost_model=AdaBoostClassifier()
ada_boost_model.fit(x_train,y_train)
ada_boost_y_pred=ada_boost_model.predict(x_test)

print("Ada Boost Classifier accuracy score", accuracy_score(y_test,ada_boost_y_pred))
print("Confusion Matrix")
print(confusion_matrix(y_test,ada_boost_y_pred))

print("Accuracy looks good but from confusion matrix it looks to be bias model for label 1")
```

## 4) Gradient Boosting Classifier

```python
grad_boost_model=GradientBoostingClassifier()
grad_boost_model.fit(x_train,y_train)
grad_boost_y_pred=grad_boost_model.predict(x_test)

print("Accuracy score of Gradient Boosting Classifier",accuracy_score(y_test,grad_boost_y_pred))
print("Confusion Matrix")
print(confusion_matrix(y_test,grad_boost_y_pred))

print("Accuracy looks to be good still from confusion matrix we could summarize it has bias model for label 1")
```
